Data Processing

```python
from sklearn.preprocessing import LabelEncoder
le= LabelEncoder()
jum['Revenue']= le.fit_transform(jum['Revenue'])
```

```python
le= LabelEncoder()
jum['VisitorType']= le.fit_transform(jum['VisitorType'])
```

```python
le= LabelEncoder()
jum['Weekend']= le.fit_transform(jum['Weekend'])
```

```python
le= LabelEncoder()
jum['Month']= le.fit_transform(jum['Month'])
```

```python
target= jum['Revenue']
jum = jum.drop(columns= ['Revenue'], axis= 1)
```

```python
jum.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler= StandardScaler()
jum= scaler.fit_transform(jum)
```

```python
jum
```

Data Modelling

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(jum, target, test_size= 0.3, random_state= 0)
```

```python
print(x_train.shape)
print(x_test.shape)
```

Using Naive Bayes Model Algorithm as it is a Supervised Learning Classifier.

```python
from sklearn.naive_bayes import GaussianNB
```

```python
nb= GaussianNB()
nb.fit(x_train,y_train)
```

```python
nb_pred= nb.predict(x_test)
nb_pred
```

MODEL EVALUATION OF NAIVE BAYES ALGORITHM

```python
from sklearn.metrics import roc_auc_score
```

```python
NB_Score = roc_auc_score(y_test,nb_pred)
```

```python
NB_Score
```

```python
y_train.value_counts(normalize=True)
```

```python
from sklearn.metrics import precision_score
```

```python
PREC_Score = precision_score(y_test,nb_pred, average=None)
```

```python
PREC_Score
```

USING DECISION TREE ALGORITHM

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
DT= DecisionTreeClassifier()
DT.fit(x_train,y_train)
```

```python
DT_pred = DT.predict(x_test)
DT_pred
```

MODEL EVALUATION USING DECISION TREE ALGORITHM

```python
DT_Score= roc_auc_score(y_test, DT_pred)
```

```python
DT_Score
```

```python
from sklearn.metrics import precision_score
```

```python
PREC_Score = precision_score(y_test,nb_pred, average=None, zero_division=1)
```

```python
PREC_Score
```
